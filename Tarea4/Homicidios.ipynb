```python
%pylab inline
import numpy.linalg as la
import csv
```

```python
#abro el archivo de homicios, gracias al codigo de Maria Camila Remolina
filas = 0
columnas = 0

with open('homicidios.1990.a.2013.csv', 'rb') as archivo:
    lector = csv.reader(archivo)
    i=0
    for fila in lector:
        if(i==0):
            for columna in fila:
                columnas+=1
        filas+=1
        i+=1

homicidios = zeros((filas-1,columnas-3))
codigo = []
municipio = []
departamento = []

with open('homicidios.1990.a.2013.csv', 'rb') as archivo:
    lector = csv.reader(archivo)
    i=0
    for fila in lector:
        if(i!=0):         
            j=0
            for columna in fila:
                if(j==0):
                    codigo.append(int(columna))
                elif(j==1):
                    municipio.append(columna)
                elif(j==2):
                    departamento.append(columna)
                else:
                    indice=(i-1,j-3)
                    if (columna=='NA' or columna=='#DIV/0!'):
                        homicidios[indice]=-1
                    else:
                        homicidios[indice]=columna
                j+=1
        i+=1
```

```python
#inicializo los valores para crear mi matriz de covarianza
anios = 24
ciudades = filas-1
valormedio = zeros(anios)
datos = zeros(ciudades)
matriz = zeros((anios, anios))
tasa = zeros((ciudades,anios))
```

```python
#saco una matriz que me tome solo los valores de las tasas
for i in range(ciudades):
   for j in range(anios):
        tasa[i,j]=homicidios[i,3*j+2]
```

```python
#saco valor medio para mi matriz de covarianza
for b in range(anios):
    valormedio[b] = mean(tasa[:,b])

#matriz de covarianza
for i in range(anios):
    for j in range(anios):
        for k in range(ciudades):
            datos[k]= (tasa[k,i] - valormedio[i])*(tasa[k,j] - valormedio[j])
            sigma =(sum(datos))/(ciudades-1) 
            matriz[i,j]=sigma
```
